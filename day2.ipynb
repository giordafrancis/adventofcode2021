```python
"""https://adventofcode.com/2021/day/2"""


# PART 1
test = """forward 5
down 5
forward 8
up 3
down 8
forward 2"""


from typing import NamedTuple

class Instruction(NamedTuple):
    direction:str
    delta:int
        
    @staticmethod
    def parse_instruction(text:str)-> NamedTuple:
        direction, delta = text.split(" ")
        return Instruction(direction, int(delta))


def creat_moves(puzzle:str)->list[Instruction]:
    moves = [
    Instruction.parse_instruction(t)
    for t in puzzle.split("\n")
    if t
]
    return moves

MOVES = creat_moves(puzzle=test)

def move_sub(moves:list[Instruction])->tuple[int, int]:
    """
    Move submarine foward or up or down in depth
    """
    x = y = 0
    for direc , delta in moves:
        if direc == 'forward':
            x += delta
        elif direc == 'up':
            y -= delta
        else:
            y += delta
    return x * y
        

assert move_sub(MOVES) == 150

with open('inputs/day2.txt') as f:
    puzzle = f.read()
    moves_puzz = creat_moves(puzzle)
    print("P1 result is ", move_sub(moves_puzz))

# PART 2

def move_sub(moves:list[Instruction])->tuple[int, int]:
    """
    Move submarine foward or up or down in depth
    track submarine aim
    """
    x = y = aim = 0
    for direc , delta in moves:
        if direc == 'forward':
            x += delta
            y += (aim * delta)
        elif direc == 'up':
            aim -= delta
        else:
            aim += delta
        #print(direc, delta)
        #print(f"x,y,aim : ({x, y, aim})")
    return x * y
        
assert move_sub(MOVES) == 900

print("P2 result is", move_sub(moves_puzz))
```
